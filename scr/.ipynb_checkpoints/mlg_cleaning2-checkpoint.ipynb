### 3.5 Correct "date" column


* Remove leading and trailing spaces

* Correct misspelling errors and unnecesary words

* Pintpoint `uncertain` records

* Drop `uncertain` values 

----

* Extract the year and month from `date`column using `datefinder` module

* Drop rows with no year/month data

```python
sharks4=sharks3.copy()
sharks5=sharks4.copy()

## Remove leading and trailing spaces (Using apply + lambda combo)
sharks5['date']=sharks5['date'].apply(lambda a: a.strip())

## Correct misspelling errors and unnecesary words
messy_terms=["Reported", "Reportd", "Mid", "Circa", "circa", "Ca."]
for i in messy_terms:
    sharks5["date"]=sharks5["date"].str.replace(i, "")

## Pintpoint uncertain records
uncertainty_terms=["Before", "before", " or ", "A.D", "B.C", "No date", "\\?", "\\/"]
for i in uncertainty_terms:
    sharks5["date"] = np.where(sharks5["date"].str.contains(i), "uncertain", sharks5["date"])

## Drop uncertain values
BB=len(sharks5)
sharks5=sharks5[sharks5["date"]!="uncertain"]

AA=len(sharks5)
print(f''' {BB-AA} rows dropped ''')

sharks5.reset_index(drop=True, inplace=True)

#sharks5[["date"]].sample(130).T
#sharks5.iloc[6202]
```

### 3.6 Correct "type" column

* Correct misspelling errors and unnecesary words by unifying the following keywords: `Boating == Boat == Boatomg`

* Pintpoint `uncertain` records (keywords: Questionable, Invalid)

* Drop `uncertain` values 

```python
sharks6=sharks5.copy()
sharks6[["type"]].value_counts()
```

```python
## Correct misspelling errors and unnecesary words
messy_terms=["Boatomg", "Boating"]
for i in messy_terms:
    sharks6["type"]=sharks6["type"].str.replace(i, "Boat")

## Pintpoint uncertain records
uncertainty_terms=["Questionable", "Invalid"]
for i in uncertainty_terms:
    sharks6["type"] = np.where(sharks6["type"].str.contains(i), "uncertain", sharks6["type"])

## Drop uncertain values
BB=len(sharks6)
sharks6=sharks6[sharks6["type"]!="uncertain"]
sharks6.reset_index(drop=True, inplace=True)

AA=len(sharks6)
print(f''' {BB-AA} rows dropped ''')

sharks6[["type"]].value_counts()
```
